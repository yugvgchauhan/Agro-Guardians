# **Entering Data**

```python
type_mapping = {
    'Stratovolcano(es)': 'Stratovolcano',
    'Shield(s)': 'Shield',
    'Lava dome(s)': 'Lava dome',
    'Caldera(s)': 'Caldera',
    'Pyroclastic cone(s)': 'Pyroclastic cone',
    'Maar(s)': 'Maar',
}

df['Type'] = df['Type'].replace(type_mapping)
type_counts = df['Type'].value_counts()

print(type_counts)
```

# **Parcing Dates**

```python
df_filtered = df[df['Last Known Eruption'] != 'Unknown']
df_filtered.info()
```

```python
import re
df_filtered_copy = df_filtered.copy()
df_filtered_copy['BCE'] = df_filtered_copy['Last Known Eruption'].apply(lambda x: int(re.sub(r'\D+', '', x)) * -1 if 'BCE' in x else None)
df_filtered_copy['CE'] = df_filtered_copy['Last Known Eruption'].apply(lambda x: int(re.sub(r'\D+', '', x)) if 'CE' in x else None)
df_filtered_copy['BCE_Period'] = df_filtered_copy['BCE'].apply(lambda x: f"{(abs(x) // 1000) * 1000 + 1} - {(abs(x) // 1000) * 1000 + 1000} BCE" if x else None)
df_filtered_copy['CE_Period'] = df_filtered_copy['CE'].apply(lambda x: f"{(abs(x) // 1000) * 1000 + 1} - {(abs(x) // 1000) * 1000 + 1000} CE" if x else None)
```

```python
df_filtered_copy['BCE_Period'] = df_filtered_copy['BCE_Period'].replace('None', np.nan)
df_filtered_copy['CE_Period'] = df_filtered_copy['CE_Period'].replace('None', np.nan)
df_filtered_copy['Period'] = df_filtered_copy['BCE_Period'].fillna(df_filtered_copy['CE_Period'])
```

```python
bce_counts = df_filtered_copy['BCE_Period'].value_counts()  # or however you calculate it

if 'nan - nan BCE' in df_filtered_copy['CE_Period'].value_counts().index:
    ce_counts = df_filtered_copy['CE_Period'].value_counts().drop(index='nan - nan BCE')
else:
    ce_counts = df_filtered_copy['CE_Period'].value_counts()

ce_counts = ce_counts.subtract(bce_counts, fill_value=0)
```
